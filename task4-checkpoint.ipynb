```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Sex')
plt.title("Passenger Count by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Pclass', y='Survived', hue='Sex', errorbar=None)  # use ci=None if errorbar not available
plt.title("Survival Rates by Gender and Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df_corr = df.drop(columns=['PassengerId'], errors='ignore')

# Compute correlation only for numeric columns
corr_matrix = df_corr.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap ", fontsize=14)
plt.show()
```

```python
pivot_table = df.pivot_table(values="Survived", index="Sex", columns="Pclass", aggfunc="mean")

plt.figure(figsize=(6,4))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Survival Heatmap (Gender vs Class)")
plt.show()
```
